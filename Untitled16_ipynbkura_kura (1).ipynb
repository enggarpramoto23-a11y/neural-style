```python
import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive, files
import os

print("Versi TensorFlow:", tf.__version__)
print("Versi TensorFlow Hub:", hub.__version__)
print("GPU Tersedia:", tf.config.list_physical_devices('GPU'))

content_path = '/content/kura-kura.jpeg'
style_path = '/content/batik pekalongan.jpeg'
output_filename = 'hasil_style_transfer_batik_pekalongan.jpg'

def load_img(path_to_img):
    max_dim = 512

    img = tf.io.read_file(path_to_img)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)

    shape = tf.cast(tf.shape(img)[:-1], tf.float32)
    long_dim = max(shape)
    scale = max_dim / long_dim

    new_shape = tf.cast(shape * scale, tf.int32)

    img = tf.image.resize(img, new_shape)
    img = img[tf.newaxis, :]
    return img

def imshow(image, title=None):
    if len(image.shape) > 3:
        image = tf.squeeze(image, axis=0)

    plt.imshow(image)
    if title:
        plt.title(title)
    plt.axis('off')

def save_img(img, filename):
    img = tf.image.convert_image_dtype(img, dtype=tf.uint8)
    img = tf.squeeze(img)
    encoded_image = tf.image.encode_jpeg(img)
    tf.io.write_file(filename, encoded_image)
    print(f"\nHasil Style Transfer telah disimpan sebagai '{filename}' di sesi Colab.")

try:
    drive.mount('/content/drive')
    print("Google Drive berhasil di-mount.")
except Exception as e:
    print(f"Gagal me-mount Google Drive: {e}")

if not os.path.exists(content_path):
    print(f"ERROR: Gambar Utama tidak ditemukan di path: {content_path}")
elif not os.path.exists(style_path):
    print(f"ERROR: Gambar Gaya tidak ditemukan di path: {style_path}")
else:
    content_image = load_img(content_path)
    style_image = load_img(style_path)

    print("\n--- Gambar berhasil dimuat ---")

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    imshow(content_image, 'Gambar Utama (Batik pekalongan)')
    plt.subplot(1, 2, 2)
    imshow(style_image, 'Gambar Gaya (kura - kura)')
    plt.show()

    try:
        hub_model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
        print("\nModel Style Transfer berhasil dimuat.")
    except Exception as e:
        print(f"ERROR: Gagal memuat model dari TF Hub. Pastikan koneksi internet stabil. {e}")
        hub_model = None

    if hub_model:
        print("Model sedang memproses Style Transfer...")
        stylized_image = hub_model(tf.constant(content_image), tf.constant(style_image))[0]
        print("Proses Style Transfer selesai.")

        save_img(stylized_image, output_filename)

        plt.figure(figsize=(8, 8))
        imshow(stylized_image, 'HASIL AKHIR (Kombinasi Gaya dan Konten)')
        plt.show()

        try:
            files.download(output_filename)
        except Exception as e:
            print(f"Gagal otomatis mengunduh. Silakan unduh manual file '{output_filename}' dari file explorer sesi Colab. {e}")
```
